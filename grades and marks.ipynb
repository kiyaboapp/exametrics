```python
import pandas as pd
from app.core.config import settings
```

```python
excel_path = r"C:\Users\droge\OneDrive\Documents\ENTRY_MBEYA_20250724_223553.xlsm"
```

```python
# Read relevant sheets
interface_df = pd.read_excel(excel_path, sheet_name="Interface")
students_df = pd.read_excel(excel_path, sheet_name="Students", usecols=["D", "E", "M", "N"])

# Rename columns for clarity
students_df.columns = ["marks", "subject_code", "exam_id", "student_global_id"]

# Get submitted_by from Interface sheet, cell A500
submitted_by = interface_df.iloc[499, 0]  # A500 (0-based index)

# Add submitted_by column
students_df["submitted_by"] = submitted_by
```

```python
# Initialize theory_marks and practical_marks columns
students_df["theory_marks"] = None
students_df["practical_marks"] = None

# Assign marks based on subject_code
students_df.loc[students_df["subject_code"].str.endswith("P"), "practical_marks"] = students_df["marks"]
students_df.loc[~students_df["subject_code"].str.endswith("P"), "theory_marks"] = students_df["marks"]

# Clean subject_code by removing "-P"
students_df["subject_code"] = students_df["subject_code"].str.replace("-P", "")
```
